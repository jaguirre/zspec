```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import readsav
```

```python
bandpassfile = 'line_cont_fitting/ftsdata/normspec_nov.sav'
```

```python
bandpass_dict = readsav(bandpassfile)
```

```python
bandpass_dict.keys()
```

```python
# The frequencies in GHz from the FTS measurement of the bandpass
nu = bandpass_dict['nu_trim']
print(nu.shape)
```

```python
# The unnormalized spectra of the 160 bands
bandpass = bandpass_dict['spec_coadd_norm']
print(bandpass.shape)
```

```python
peak = bandpass_dict['peak']
```

```python
plt.figure(figsize=(16,5))
for i in np.arange(160):
    plt.plot(nu, bandpass[:,i] / peak[i])
plt.title('All normalized bandpasses')
plt.xlim([180, 312])
```

```python
plt.figure(figsize=(16,5))
for i in np.arange(0, 160, 10):
    plt.plot(nu, bandpass[:,i] / peak[i])
plt.title('Selected normalized bandpasses')
plt.xlim([180, 312])
```
